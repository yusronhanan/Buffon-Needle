

*   N = number of needle
*   l = needle's length. l < d
*   d = lebar garis
*   Q = total jarum yang menyentuh garis
*   0 <= x <= d dimana x = l*cos(theta)
*   0 <= theta <= 90
*   Hitung nilai W = ((2l)/d *) (N\Q) 
                    = 2LN / dq













```
import random 
import matplotlib.pyplot as plt
import math
import numpy as np

l = 4.5
d = 6
loop = 10
ex = []
ye = []
# number of needle of 100 - 10.100
N = 100
while N <= 10100:
    ptot = 0
    for i in range(loop):
        Q = 0
        for j in range(N):
            theta = random.uniform(0,math.pi/2)
            x = random.uniform(0,d)
            if x <= l*math.cos(theta):
                Q = Q + 1
        p = (2*l*N)/(d*Q)
        ptot = ptot + p
    ex.append(N)
    ye.append(ptot/loop)
    N = N + 100

plt.plot(ex,ye)
plt.show()
```
